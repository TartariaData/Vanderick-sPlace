```python
# links to websites
Link="https://docs.google.com/spreadsheets/d/e/2PACX-1vSPVJIoDhupNyUpYD65FSHt6kMmF_KaxE2RD6CV1FOMP4vVw7WfKQMqpqo0ieD9vhMAmYP1NpBmw0od/pubhtml?gid=658985149&single=true"
# scrapping into LIST of Dataframes

import pandas as pd

List=pd.read_html(Link,header=0,flavor='bs4')


# getting the Dataframe from list
data=List[0]

#no spaces in column names
data.columns=data.columns.str.replace(r'\s','',regex=True)

# Eliminar las filas en los índices 0, 1, 2 y 3
data = data.drop([0,1,2,3,4,5,6,7])


# dropping
toDrop=['Unnamed:0','Unnamed:1','Unnamed:4','Unnamed:6','Unnamed:8','Unnamed:10','Unnamed:12','Unnamed:14','Unnamed:15']
data.drop(columns=toDrop,inplace=True)

#renaming
newDataNames={'Unnamed:2':'Country','Unnamed:3':'Human_Development','Unnamed:5':'Life_expectancy_at_birth','Unnamed:7':'Expected_years_of_schooling','Unnamed:9':'Mean_years_of_schooling', 'Unnamed:11':'GNI_per_capita','Unnamed:13':'GNI_per_capita_rank_minus_HDI_rank'}
data.rename(columns=newDataNames,inplace=True)
```
