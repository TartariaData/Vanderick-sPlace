```python
#PASO N°1: Integrar archivos

# Creamos un repositorio con el archivo appendData
Repositorio='https://github.com/TartariaData/TAREAS/raw/main/carpetaIntegrar/'

# Creamos nombres a nuestros archivos
levels=range(1,5)
names=['level'+str(level)+'.xlsx' for level in levels]

# Creamos los links de archivos
# Dentro de un listado de data frames
Links=[Repositorio+fn for fn in names]
Links
```

```python
# Instalamos bibliotecas utiles
!pip install openpyxl
!pip install xlrd
!pip install numpy
```

```python
# Creamos un listado con todos los Data Frames (cada excel representa uno)
import pandas as pd

DataFrames=[pd.read_excel(link) for link in Links]
```

```python
# Creamos una formula para observar la información de cada data frame
def levelx(DataFrames, x):

  if x < 1 or x > len(DataFrames):
    raise ValueError(f"El índice 'x' debe estar entre 1 y {len(DataFrames)}")

  levelx = DataFrames[x - 1]
  print(f"Información del nivel {x}:")
  levelx.info()
  print(f" ")

# Usamos nuestra formula
for i in range(1, 5):
  levelx(DataFrames, i)
```

```python
# Observamos que el archivo de level1 necesita renombrar los nombres de sus columnas y formatearse
level1 = DataFrames[0]
level1Cleaned= level1
level1Cleaned = level1Cleaned.rename(columns={'Unnamed: 0': 'HDI rank', 'Table 1. Human Development Index and its components ': 'Country', 'Unnamed: 2': 'Human Development Index (HDI) '
, 'Unnamed: 4': 'Life expectancy at birth', 'Unnamed: 6': 'Expected years of schooling', 'Unnamed: 8': 'Mean years of schooling', 'Unnamed: 10': 'Gross national income (GNI) per capita'
, 'Unnamed: 12': 'GNI per capita rank minus HDI rank', 'Unnamed: 14': 'HDI rank.1'})
level1Cleaned.head()
```

```python
# Eliminamos las primeras filas que contienen datos innecesarios
level1Cleaned = level1Cleaned.drop(index=level1Cleaned.index[:6])

# Convertir las columnas numéricas a tipo numérico
import numpy as np

columnas_numericas = ['Human Development Index (HDI) ', 'Life expectancy at birth',
                      'Expected years of schooling', 'Mean years of schooling',
                      'Gross national income (GNI) per capita','GNI per capita rank minus HDI rank', 'HDI rank.1']

level1Cleaned[columnas_numericas] = level1Cleaned[columnas_numericas].apply(pd.to_numeric)

# Verificar la información actualizada del DataFrame
print(level1Cleaned.info())

level1= level1Cleaned
DataFrames[0]= level1
```

```python
# Creamos una columna vacia y vamos a pegarle los nombres necesarios
# Lo que queremos es crear un data frame juntando todos los datas frame que tenemos
ColumnNames=[]
for df in DataFrames:
    ColumnNames.append(set(df.columns))

# Checamos cuantos columnas hay en cada dataframe
[len(cols) for cols in ColumnNames]
```

```python
# Vamos a observar los detalles de las columnas y ver las columnas comunes en todos los data frame
print(f"Columnas Comunes:")
commonColumns=set.intersection(*ColumnNames)
len(commonColumns),commonColumns
```

```python
# Observamos las columnas distintas al restar todas las columnas menos las columnas comunes
print(f"Columnas Distintas:")
set.union(*ColumnNames)-commonColumns
```

```python
# Un listado de Data Frames con todas las columnas comunes
listaDataIntegrada=[df.loc[:,list(commonColumns)] for df in DataFrames]

# Creamos el Data Frame integrado y vemos algunos de sus valores
DataIntegrada=pd.concat(listaDataIntegrada)
DataIntegrada.head()
```

```python
# Eliminar filas con al menos 1 valores nulos al ser filas que no contienen paises ni datos
umbral = 1
DataIntegrada =DataIntegrada.dropna(axis=0, thresh=DataIntegrada.shape[1] - umbral)

# Reiniciar la numeración de las filas
DataIntegrada = DataIntegrada.reset_index(drop=True)

# Ordenaremos nuestro Data Frame, nuestro criterios de columna Country y HDI Rank
DataIntegrada.set_index(['Country','HDI rank'],inplace=True)

# Ordenamos alfabeticamente nuestras columnas
# empezamos desde el axis 1 que es donde comuienzan las columnas
DataIntegrada.sort_index(axis=1,inplace=True)

# Regresamos nuestro criterios a columnas:
DataIntegrada.reset_index(inplace=True)
DataIntegrada.head()
```

```python
# Reescribimos los paises en mayusculas
DataIntegrada.Country=DataIntegrada.Country.str.upper()

# Por ultimo importamos nuestro archivo integrado como un archivo csv
import os
DataIntegrada.to_csv("DataIntegrada.csv",index=False)
DataIntegrada
```

```python
# PASO 2: Merge de archivos
# Importamos nuestra data
import pandas as pd

countryCODESLink='https://github.com/TartariaData/TAREAS/raw/main/countryCODES.xlsx'
countryCODES=pd.read_excel(countryCODESLink)

print(f"Observamos información:")
countryCODES.info()
print(f" ")
print(f"Observamos algunos datos:")
countryCODES.head()
```

```python
# Visualizamos nuestro maximo merge (que corresponde al menor resultado)
countryCODES.shape,DataIntegrada.shape
```

```python
# Instalaremos esa biblioteca el thefuzz
!pip install thefuzz
from thefuzz import process as fz

# Agregaremos los codigos de paises (countryCODES) a nuestro data integrado anteriormente
countryCODES.Country=countryCODES.Country.str.upper()
countryCODES.merge(DataIntegrada,left_on='Country',right_on='Country')
```

```python
# Identificamos nombres similares entre ambas datas
onlyData=set(DataIntegrada.Country)-set(countryCODES.Country)
onlyCODES=set(countryCODES.Country)-set(DataIntegrada.Country)

[(f,fz.extractOne(f, onlyCODES)) for f in sorted(onlyData)]

# Tras ello guardamos los valores en un directorio
changes={f:fz.extractOne(f, onlyCODES)[0]
                 for f in sorted(onlyData)
                 if fz.extractOne(f, onlyCODES)[1] >=87}

# Hacemos el cambio , cuyo representa unir los cambios dentro de nuestra data
DataIntegrada.replace(to_replace={'Country':changes},inplace=True)
```

```python
# Volvemos a hacer lo mismo en busca de valores aun similares
onlyData=set(DataIntegrada.Country)-set(countryCODES.Country)
onlyCODES=set(countryCODES.Country)-set(DataIntegrada.Country)

[(f,fz.extractOne(f, onlyCODES)) for f in sorted(onlyData)]
```

```python
# Observamos que ya no hay nombres de paises similares, entonces hacemos el merge
DataIntegrada_CODES=countryCODES.merge(DataIntegrada,left_on='Country',right_on='Country')

# Guardamos nuestro data
DataIntegrada_CODES.to_csv("DataIntegrada_CODES.csv", index=False)

# Observamos el resultado final
DataIntegrada_CODES
```

```python
#PASO 3: Crear un Reporte mensual de fallecidos
import pandas as pd
fallecidosCOVIDLink=pd.read_csv("https://raw.github.com/TartariaData/TAREAS/main/fallecidosCOVID.csv", sep=";", header=None)
fallecidosCOVID = fallecidosCOVIDLink
print(f" ")

# Extrae los nombres de columna de la primera fila
nombres_columna = fallecidosCOVID.iloc[0].values.tolist()
print(f" ")

# Asigna los nombres de columna al DataFrame, usando comillas dobles si es necesario
fallecidosCOVID.columns = [nombre.replace(';', '"') if ';' in nombre else nombre for nombre in nombres_columna]
print(f"Observamos información:")
fallecidosCOVID.info()
print(f" ")
print(f"Observamos algunos datos:")
fallecidosCOVID.head()
```

```python
# Limpiaremos la data eliminando la primera fila (esta extra)
fallecidosCOVID = fallecidosCOVID.drop(0)

# Reiniciar la numeración de las filas
fallecidosCOVID = fallecidosCOVID.reset_index(drop=True)
fallecidosCOVID
```

```python
# Formateamos la columna de interes a tipo fecha
fallecidosCOVID["FECHA_FALLECIMIENTO_NUMERICO"] = pd.to_numeric(fallecidosCOVID["FECHA_FALLECIMIENTO"], errors='coerce')
fallecidosCOVID["FECHA_FALLECIMIENTO_FORMATO"] = pd.to_datetime(fallecidosCOVID["FECHA_FALLECIMIENTO"], format='%Y%m%d')
fallecidosCOVID.FECHA_FALLECIMIENTO_FORMATO
```

```python
# Creamos una copia
fallecidosCOVID_original = fallecidosCOVID.copy()

# Creamos una nueva columna para contar las muertes diarias
fallecidosCOVID['MUERTES_DIARIAS'] = 1

# Contamos las muertes diarias
fallecidos_diarios = fallecidosCOVID.groupby('FECHA_FALLECIMIENTO_FORMATO')['MUERTES_DIARIAS'].sum().reset_index()

# Extraemos los meses y años
fallecidos_diarios['Mes'] = fallecidos_diarios['FECHA_FALLECIMIENTO_FORMATO'].dt.month
fallecidos_diarios['Año'] = fallecidos_diarios['FECHA_FALLECIMIENTO_FORMATO'].dt.year

# Creamos nuestro reporte global de muertes por mes
reporte_mensual_global = fallecidos_diarios.groupby(['Mes', 'Año'])['MUERTES_DIARIAS'].sum().reset_index().rename(columns={'MUERTES_DIARIAS': 'Cantidad de Fallecidos'})

# Observamos reporte
reporte_mensual_global.head()
```

```python
# Guardamos resultados con su nombre final
ReporteMensual_Año=reporte_mensual_global

# Guardamos nuestro data
ReporteMensual_Año.to_csv("ReporteMensual_Año.csv", index=False)

# Mostrar el reporte mensual por mes
ReporteMensual_Año
```

```python
#PASO 4: Crear un Reporte mensual de fallecidos que especifique el tipo de muerte
import pandas as pd

# Importamos nuestro archivo nuevamente
fallecidosCOVIDLink2 = pd.read_csv("https://raw.github.com/TartariaData/TAREAS/main/fallecidosCOVID.csv", sep=";", header=None)
fallecidosCOVID2 = fallecidosCOVIDLink2

# Extraemos las columnas
nombres_columna = fallecidosCOVID2.iloc[0].values.tolist()

# Reemplazamos los separadores
fallecidosCOVID2.columns = [nombre.replace(';', '"') if ';' in nombre else nombre for nombre in nombres_columna]

# Limpiaremos la data eliminando la primera fila (esta extra)
fallecidosCOVID2 = fallecidosCOVID2.drop(0)

# Reiniciar la numeración de las filas
fallecidosCOVID2 = fallecidosCOVID2.reset_index(drop=True)
```

```python
#Observamos nuestra data
print(f'Observamos e identificamos el nombre de la columna de interes "tipo de muerte"')
print(f' ')
fallecidosCOVID2.columns
#Su nombre es 'CLASIFICACION_DEF
```

```python
# Formateamos la columna de interes a tipo fecha
fallecidosCOVID2["FECHA_FALLECIMIENTO_NUMERICO"] = pd.to_numeric(fallecidosCOVID2["FECHA_FALLECIMIENTO"], errors='coerce')
fallecidosCOVID2["FECHA_FALLECIMIENTO_FORMATO"] = pd.to_datetime(fallecidosCOVID2["FECHA_FALLECIMIENTO"], format='%Y%m%d')
fallecidosCOVID2.FECHA_FALLECIMIENTO_FORMATO

# Accedemos a los atributos "año" y "mes" para el procesamiento
fallecidosCOVID2["Año"] = fallecidosCOVID["FECHA_FALLECIMIENTO_FORMATO"].dt.year
fallecidosCOVID2["Mes"] = fallecidosCOVID["FECHA_FALLECIMIENTO_FORMATO"].dt.month

#Creamos el reporte mensual en base al tipo de muerte
reporte_mensual_con_tipo_muerte = fallecidosCOVID2.groupby(['Mes', 'Año', 'CLASIFICACION_DEF'])['FECHA_FALLECIMIENTO'].count().unstack().fillna(0)
reporte_mensual_con_tipo_muerte = reporte_mensual_con_tipo_muerte.reset_index()
reporte_mensual_con_tipo_muerte.rename(columns={'FECHA_FALLECIMIENTO_FORMATO': 'Fallecidos'}, inplace=True)
reporte_mensual_con_tipo_muerte.head(5)
```

```python
# Guardamos el resultado final
ReporteMensual_TipoMuerte=reporte_mensual_con_tipo_muerte

# Guardamos nuestro data
ReporteMensual_TipoMuerte.to_csv("ReporteMensual_TipoMuerte.csv", index=False)

# Mostrar el reporte mensual por mes
ReporteMensual_TipoMuerte
```
