```python
#Importamos la base de datos en formato CSV
import pandas as pd

linkToFile='https://docs.google.com/spreadsheets/d/e/2PACX-1vSPVJIoDhupNyUpYD65FSHt6kMmF_KaxE2RD6CV1FOMP4vVw7WfKQMqpqo0ieD9vhMAmYP1NpBmw0od/pub?gid=658985149&single=true&output=csv'
human=pd.read_csv(linkToFile)
```

```python
#Visualizamos la tabla para ver cómo se importó
human
```

```python
# Eliminar filas con al menos 13 valores nulos
umbral = 13
human_clean = human.dropna(axis=0, thresh=human.shape[1] - umbral)
human_clean
```

```python
#Actualizamos el valor de human por human_clean
human = human_clean
```

```python
# Eliminar las cuatro primeras filas que contienen datos innecesarios
human_clean = human.drop(index=human.index[:4])

human_clean
```

```python
#Actualizamos el valor de human por human_clean
human = human_clean
```

```python
# Contar el número de valores nulos en cada columna
null_counts = human.isnull().sum()

# Obtener los índices de las columnas que tienen al menos 5 valores nulos
columns_to_drop = null_counts[null_counts >= 5].index

# Eliminar las columnas con al menos 5 valores nulos
human_clean = human.drop(columns=columns_to_drop)
human_clean
```

```python
#Actualizamos el valor de human por human_clean
human = human_clean
```

```python
# Renombrar las columnas
new_names = {
    'Table 1. Human Development Index and its components': 'Country',
    'Unnamed: 2': 'Human development index (HDI)',
    'Unnamed: 4': 'Life expectancy at birth (years)',
    'Unnamed: 6': 'Expected years of schooling (years)',
    'Unnamed: 8': 'Mean years of schooling (years)',
    'Unnamed: 10': 'Gross national income (GNI per capita)',
    'Unnamed: 12': 'GNI per capita rank minus HDI rank',
    'Unnamed: 14': 'HDI rank'
}
human_clean = human_clean.rename(columns=new_names)
human_clean
```

```python
#Actualizamos el valor de human por human_clean
human = human_clean
```

```python
# Reiniciar la numeración de las filas
human_clean = human_clean.reset_index(drop=True)
#Actualizamos el valor de human por human_clean
human = human_clean
human
```

```python
# Extraer el valor de la fila 64 para comprobar si se han limpiado los datos adecuadamente
valor_fila_64 = human_clean.iloc[64]
# Mostrar el valor
print("El valor de la fila 64  es:", valor_fila_64)
```

```python
# Buscar la fila donde el país es "Very high human development" porque desde ahí se borrará todo lo que no es informaicón de país
fila = human_clean.loc[human_clean['Country'] == 'Very high human development']

# Obtener el número de índice de la fila
numero_fila = fila.index[0]

# Mostrar el número de índice
print("El número de índice de la fila donde el país es 'Very high human development' es:", numero_fila)
```

```python
# Eliminar filas desde la 195 en adelante, porque desde esa fila ya no son países
human_clean = human_clean.drop(index=human_clean.index[195:])
#Actualizamos el valor de human por human_clean
human = human_clean
human
```

```python
#Revisamos qué tipo de datos hay en las columnas
human.info()
```

```python
# Identificar las columnas numéricas
columnas_numericas = ['Human development index (HDI)', 'Life expectancy at birth (years)',
                      'Expected years of schooling (years)', 'Mean years of schooling (years)',
                      'Gross national income (GNI per capita)', 'GNI per capita rank minus HDI rank',
                      'HDI rank']

# Iterar sobre las columnas numéricas y buscar símbolos y espacios
for columna in columnas_numericas:
    valores_con_simbolos = human_clean[columna].str.contains(r'[^\d.\s-]', na=False)
    valores_con_espacios = human_clean[columna].str.contains(r'\s', na=False)

    print(f"Valores con símbolos en la columna '{columna}':")
    print(human_clean[columna][valores_con_simbolos])
    print(f"Valores con espacios en la columna '{columna}':")
    print(human_clean[columna][valores_con_espacios])
    print()
```

```python
# Eliminar las comas de la columna "Gross national income (GNI per capita)"
human_clean['Gross national income (GNI per capita)'] = human_clean['Gross national income (GNI per capita)'].str.replace(',', '')

#Actualizamos el valor de human por human_clean
human = human_clean
human
```

```python
#Revisamos qué tipo de datos hay en las columnas
human.info()
```

```python
# Identificar las columnas numéricas
columnas_numericas = ['Human development index (HDI)', 'Life expectancy at birth (years)',
                      'Expected years of schooling (years)', 'Mean years of schooling (years)',
                      'Gross national income (GNI per capita)', 'GNI per capita rank minus HDI rank',
                      'HDI rank']

# Iterar sobre las columnas numéricas y buscar valores no numéricos
for columna in columnas_numericas:
    valores_no_numericos = human_clean[~human_clean[columna].str.match(r'^-?\d*\.?\d*$')]

    print(f"Valores no numéricos en la columna '{columna}':")
    print(valores_no_numericos)
    print()
```

```python
import numpy as np

# Reemplazar los valores no numéricos con NaN
human_clean.replace('..', np.nan, inplace=True)

# Convertir las columnas numéricas a tipo numérico
columnas_numericas = ['Human development index (HDI)', 'Life expectancy at birth (years)',
                      'Expected years of schooling (years)', 'Mean years of schooling (years)',
                      'Gross national income (GNI per capita)', 'GNI per capita rank minus HDI rank',
                      'HDI rank']

human_clean[columnas_numericas] = human_clean[columnas_numericas].apply(pd.to_numeric)

# Verificar la información actualizada del DataFrame
print(human_clean.info())
```
