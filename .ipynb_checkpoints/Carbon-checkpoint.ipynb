```python
#Leemos
import pandas as pd

linkToFile='https://github.com/CienciaDeDatosEspacial/code_and_data/raw/main/data/carbonEmi_downloaded.csv'
carbon=pd.read_csv(linkToFile)
```

```python
#Visualisamos
carbon
```

```python
#Observamos la información 
carbon.info()
```

```python
#Frecuencia
carbon.region.value_counts()
```

```python
# Frecuencia
carbon.date_of_information.value_counts()
```

```python
#1.Conservamos las columnas name, value, date_of_information and region
carbonCopia1=carbon.drop(columns=['slug'])
carbonCopia1
```

```python
#2. Cambiamos el nombre data_of_information a carbon_date
carbonCopia2=carbonCopia1.rename(columns={'date_of_information':'carbon_date'})
carbonCopia2
```

```python
#3. Sin espacios al inicio ni al final
carbonCopia2['name']=carbonCopia2.name.str.strip()
carbonCopia2
```

```python
#4. Detección de simbolos, celdas con caracteres no numericos
carbonCopia2.value.str.contains(pat=r'\D',
                           regex=True)
```

```python
#5. Nombres de columnas sin espacios
carbonCopia2['name']=carbonCopia2.name.str.replace(pat='\s',repl='',regex=True)
carbonCopia2
```

```python
#6. Desacernos de los valores hallados en el paso 5
#A que se refiere?
```

```python
#7. Manten solo el valor del año en "carbon_date"
#carbon2.carbon_date.str.extract(pat=r'([2019]...)', expand=True)

carbonCopia2['carbon_date']=carbonCopia2.carbon_date.str.extract(pat=r'(^\w+|.$)', expand=True)
carbonCopia2
```

```python
#Archivo limpio
carbonCleaned=carbonCopia2
carbonCleaned
```

```python
import os
carbonCleaned.to_csv(os.path.join("Vanderick'sPlace","carbonCleaned.csv"),index=False)
```

```python
#Exercise 2
from IPython.display import IFrame  
ciaLink2="https://www.cia.gov/the-world-factbook/field/revenue-from-forest-resources/country-comparison" 
IFrame(ciaLink2, width=900, height=900)
```

```python
# read web table into pandas DF
import pandas as pd

forestDFs=pd.read_html(ciaLink2, # link
                       header=0, # where is the header?
                       flavor='bs4')
```

```python
# make a copy
forest=forestDFs[0].copy()
```

```python
# here it is
forest
```

```python
#1. Replace '%' by 'pct'
forest1=forest.rename(columns={'% of GDP':'pct of GDP'})
forest1
forest.columns.str.replace('%','pct')
```

```python
forest.columns=forest.columns.str.replace('%','pct')
forest
```

```python
#2. Keep the columns Country, pct of GDP, and Date of Information
forest2=forest1.drop(columns=['Rank'])
forest2
```

```python
#3. Change the column name Date of Information to forest_date.
forest3=forest2.rename(columns={'Date of Information':'forest_date'})
forest3
```

```python
#4. Make sure there are no spaces as part of the column names
forest3['Country']=forest3.Country.str.replace(pat='\s',repl='',regex=True)
forest3
```

```python
#5. Make sure the cells with text does not have neither trailing nor leading spaces
forest3['Country']=forest3.Country.str.strip()
forest3
```

```python
#6. Keep only the year value in the column forest_date.
forest3['forest_date']=forest3.forest_date.str.extract(pat=r'(^\w+|.$)', expand=True)
forest3
```

```python
forestCleaned=forest3
forestCleaned
```

```python
#mi carpeta si existe, pero sale que no :(
import os
forestCleaned.to_csv(os.path.join("TAREAS","forestCleaned.csv"),index=False)
```
